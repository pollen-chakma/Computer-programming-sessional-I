## Arithmatic Operator
Python operators are fundamental for performing mathematical calculations. Arithmetic operators are symbols used to perform mathematical operations on numerical values. Arithmetic operators include addition (+), subtraction (-), multiplication (*), division (/), and modulus (%)

![](ArithmaticOperator.png)

# Conditional Statement in Python
Conditional Statement in Python perform different computations or actions depending on whether a specific Boolean constraint evaluates to true or false. Conditional statements are handled by IF and Elif statements in Python.
Type of condition statement in Python:
1. If statement
2. If Else statement
3. Elif statement
4. Nested if statement.
  
* Conditional statements are also called decision-making statements. We use those statements while we want to execute a block of code when the given condition is true or false.
* n answer have n-1 condition

## if-else statement

![](if-else-statement.png)

The if-else statement is used to code the same way you would use it in the English language. The syntax for the if-else statement is:

if(condition):
      
      statement block // for when condition is TRUE
else:
      
      statement block //for when condition is FALSE

```python
#if else

a = int(input())
b = int(input())

if(a<b): # condition can be written with or without bracket
    print("a is smaller than b")
else:
    print("a is greater than b")
```

```python
# Depending on the river discharge, write a python program to categorizes river flow into three different classes for flood forecasting. 
#These three classifications are Low(Discharge <50cumsec) , Medium(50<=discharge<=150 cumsec), High(Discharge>150 cumsec)(. This can help in flood forecasting or water resource management.

discharge = float(input("Input river discharge in cumsec"))
if discharge < 50:
    print("Low flow")
elif discharge >= 50 and discharge <= 150:
    print("Medium flow")
else:
    print("High flow")
```

```python
# Depending on the river discharge, write a python program to categorizes river flow into three different classes for flood forecasting. 
#These three classifications are Low(Discharge <50cumsec) , Medium(50<=discharge<=150 cumsec), High(Discharge>150 cumsec)(. This can help in flood forecasting or water resource management.

# write the same using nested if-else statement

discharge = float(input("Input river discharge in cumsec"))
if discharge <= 150:
    if discharge >=50 and discharge < 150:
        print("Medium flow")
    else: 
        print("Low flow")
else:
    print("High flow")
```

Course Grading System
In a course, grades are assigned based on a student’s marks according to the following criteria:

* Marks   Range Grade
* 90–100	A+
* 80–89	    A
* 70–79	    A-
* 60–69	    B+
* 50–59	    B
* 40–49	    C
* Below 40  F

Additional Requirements:

* If the marks are not within the range of 0–100, the program should print an error message: "Invalid marks. Please enter a value between 0 and 100."
* Use a nested if-else statement to determine the grade.
* The script should take the marks as input from the user and output the grade.


Task: Write a Python script of a course grading system that prompts the user to input the marks.

```python
# Course Grading System

# Prompt the user to input marks
marks = float(input("Enter the marks (0-100): "))

# Validate marks range
if marks >= 0 and marks <= 100:
    if marks >= 90:
        print( "A+")
    elif marks >= 80:
        print( "A")
    elif marks >= 70:
        print( "A-")
    elif marks >= 60:
        print( "B+")
    elif marks >= 50:
        print( "B")
    elif marks >= 40:
        print( "C")
    else:
        print( "F")
else:
    print("Invalid marks. Please enter a value between 0 and 100.")
```

```python
# Course Grading System(Alternative solution)

# Prompt the user to input marks
marks = float(input("Enter the marks (0-100): "))

# Validate marks range
if marks < 0 or marks > 100:
    print("Invalid marks. Please enter a value between 0 and 100.")
else:
    # Determine the grade using nested if-else
    if marks >= 90:
        print( "A+")
    elif marks >= 80:
        print( "A")
    elif marks >= 70:
        print( "A-")
    elif marks >= 60:
        print( "B+")
    elif marks >= 50:
        print( "B")
    elif marks >= 40:
        print( "C")
    else:
        print( "F")
```
