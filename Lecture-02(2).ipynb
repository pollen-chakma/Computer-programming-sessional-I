## 1.Variables


#### What is variable?
A Python variable is a reserved memory location to store values. In other words, a variable in a python program
gives data to the computer for processing.
We can consider it as a container that can store value or take value directly or from keyboard.

A = 5, B = 6
Here, A and B are two variables.
#### Rules for Python variables:
1. A variable name must start with a letter or the underscore character. (a, name, _name, var)
2. A variable name cannot start with a number. (Iname, 1 _name) (numl, namel, name_ 1)
3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and ) (First-name) (First name)
4. Variable names are case-sensitive (age, Age and AGE are three different variables) (Name, name)

#### Some important tips:

* Declare a meaningful variable name instead of a single letter.
* Conscious about the case of the letters. (Both declaration and use)
* Variable name should be related to the problem/ program.

```python
#A = 5, B = 6
#Here, A and B are two variables.

A = 2 #Here = is not equal sign. Its called assignment operator
B = 10
Name = 'Python'
name = "Python" #Both single and double quotation can be used for string

print(A)
print(B)
print(Name)
print(name)
```
