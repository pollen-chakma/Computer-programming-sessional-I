# Introduction to loop

Loops are control flow statements that allow us to execute a block of code multiple times. They are essential for automating repetitive tasks, processing data efficiently, and reducing redundant code.

* Python provides two main types of loops:

#### 1. for loop - Iterates over a sequence (such as a list, tuple, dictionary, or string).

#### 2. while loop - Repeats as long as a condition remains true.

#### 3. Nested Loop - One or more loop inside another for loop or while loop



We can solve this using loop.
* If you need to print any pattern then you need loop.
* If you want to access a lot of elements then you need loop.
* If you want to control the data structures then you need loop.

### A loop contains 3 parts

1. Initialization 
2. Condition
3. Increment/Decrement

# Exercise: if we want to print Thank you 10 times side by side[having a space gap]

```python
#if we want to print side by side[having a space gap]
# Take user input
number = 0  

# Condition of the while loop
while number <=9 :  
    print("Thank you", end=" ")
    # Increment the value of the variable "number by 1"
    number = number+1
```

```python
# print the even numbers from 1 to n using while loop

n = int(input("Enter a number: "))  # Taking user input
i = 2  # Start from the first even number

while i <= n:
    print(i, end=" ")  # Print even number
    i += 2  # Increment by 2 to get the next even number
```

```python
# alternative solution using conditional and while loop

n=int(input())
i=1
while i<=n:
    if(i%2==0):
        print(i,end=" ")
    i+=1
```

```python
# print the even numbers from 1 to n using for loop

n = int(input("Enter a number: "))  # Taking user input

for i in range(2, n + 1, 2):  # Start from 2, increment by 2
    print(i, end=" ")
```

```python
#alternative solution using conditional statement and for loop

n = int(input("Enter a number: "))  # Taking user input

for i in range(1, n + 1):  # Loop from 1 to n
    if i % 2 == 0:  # Check if the number is even
        print(i, end=" ")
```

# Nested Loop

A nested loop is a loop inside another loop. The "inner loop" runs completely every time the "outer loop" runs once. Nested loops are useful when you need to perform repetitive tasks that require multiple levels of iteration, such as iterating over two-dimensional structures like matrices or grids.

```python
# Outer loop runs 5 times

for i in range(5):
    # Inner loop runs from 1 to 10
    for j in range(1, 11):
        print(j, end=" ")
    print()  # Moves to the next line after each outer loop iteration
```
