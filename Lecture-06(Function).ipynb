## What are functions in python

Python Functions is **a block of statements** that return the specific task. The idea is to put **some commonly or repeatedly done tasks** together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.

or A function is **a block of organized, reusable code** that is used to perform a single, related action. Functions
provide better modularity for your application and a high degree of code reusing.

e.g. **print(), type()** etc. 

**Some Benefits of Using Functions:**

* Increase Code Readability 
* Increase Code Reusability

Above shown is a function definition that consists of the following components:
1. Keyword **def** that marks the start ot the function header.
2. A function name to uniquely identify the function. Function naming follows the same rules of writing **variable name** 
3. Parameters (arguments) through which we pass values to a function. They are optional.

6. An optional **return statement** to return a value from the function.

## Python return statement

A return statement is used **to end the execution of the function call and it “returns” the value of the expression following the return keyword** to the caller. The statements after the return statements are not executed. If the return statement is without any expression, then the special value **None** is returned. A return statement is overall used to invoke a function so that the passed statements can be executed.

**When the return statement is executed, the function terminates and the specified value is returned to the caller.** If no value is specified, the function returns **None** by default.

N.B. Return statement can not be used outside the function.

### No Argument nothing Return

1. No Argument means it does not take anything outside the function. It means the input is taken inside the function.
2. Nothing return means, after processing the output does not return to anywhere. It means the print statements are inside the function.

```python
def addition():
    a,b = map(int, input("Enter Two numbers:") .split ())
    sum = a + b
    print("The Sum is = ",sum)
```

## Exercise: 01

**Groundwater Recharge Estimation:**

Question: Write a python script for creating a **function for Groundwater Recharge Estimation** with these following conditions:

* Rainfall > 2000mm → Excellent Recharge Potential
* Rainfall > 1000mm → Good Recharge Potential
* Rainfall > 500mm → Moderate Recharge Potential
* Rainfall ≤ 500mm → Poor Recharge Potential - Risk of Groundwater Depletion

```python
def groundwater_recharge(rainfall):
    """Determines groundwater recharge category based on annual rainfall (in mm)."""
    
    if rainfall > 2000:
        return "Excellent Recharge Potential"
    elif rainfall > 1000:
        return "Good Recharge Potential"
    elif rainfall > 500:
        return "Moderate Recharge Potential"
    else:
        return "Poor Recharge Potential - Risk of Groundwater Depletion"
```

```python
rain = float(input("Enter annual rainfall (mm): "))
print(groundwater_recharge(rain))
```
