## What is Data Sturucture?

Data structures are a way of organizing and storing data so that they can be accessed and worked with efficiently.
They define the relationship between the data, and the operations that can be performed on the data. 

## Different types of data structure used in python:

Data structures in python can be classified into two major categories. One is primitive data structures and another one is non-primitive data structures.

* **Primitive Data Structures**
1. Integers
2. Float
3. Strings
4. Boolean
   
* **Non-Primitive Data Structures**
1. Arrays
2. **Lists**
3. **Tuples**
4. **Dictionary**
5. **Sets**
6. Files

Also there are some advanced data structures available in python; **linked lists, stacks, queues, heaps, trees, tries, graphs, and hash tables**.

## Why Data Sturucture?

Data structures in Python are used to organize, store, and manage data efficiently. They help in performing various operations such as searching, sorting, inserting, and deleting data with optimized performance. Here’s why data structures are essential in Python:

1. **Efficient Data Management**
* Data structures like lists, dictionaries, sets, and tuples allow efficient storage and retrieval of data.
  * Data Structures are used to store a collection of related data.
  * Data structures provide a means to manage large amounts of data efficiently for uses such as large databases and internet indexing services.
  * Data structures are key to designing efficient algorithms(Problem solving procedure).
  * Data structures allow information storage on hard disks.
  * Data structures allows safe storage of information on a computer.
* They help in structuring data logically, making it easier to manipulate.

2. **Optimized Performance**
* Using the right data structure reduces time complexity. For example:
* Lists allow fast appending and iteration.
* Dictionaries provide quick lookups using hash tables.
* Sets eliminate duplicate values and support fast membership tests.

3. **Facilitates Complex Operations**
* Advanced structures like queues, stacks, trees, and graphs help in solving complex problems like graph traversal, scheduling, and network routing.

4. **Memory Efficiency**
* Proper selection of data structures helps in minimizing memory usage.
* Tuples consume less memory compared to lists as they are immutable.

5. **Data Integrity and Security**
* Using immutable data structures like tuples ensures that data remains unchanged, preventing accidental modifications.

6. **Flexibility in Programming**
* Data structures support different paradigms like object-oriented programming and functional programming, making Python more powerful and adaptable.

7. **Real-world Applications**
* Used in databases (hash tables for indexing).
* Used in artificial intelligence (graphs for neural networks).
* Used in web development (caching with dictionaries and sets).
* Used in machine learning (numpy arrays for numerical data processing).

## Some speficific use case scenario of different Data Sturucture?

* Linked lists are purpose-built for **inserting and deleting data** but **only offer sequential access to this data, which means searching and sorting will be problematic.**
* Similar to a list, tuples can be used for **outputting either an entire tuple or individual elements, but they use less memory space and do not **allow sorting, adding, replacing or deleting elements.**
* Sets are intentionally **designed to be limited but excel in checking for a value’s existence and avoiding duplicates in a set.**
* Dictionaries allow **data to be collected in key-value pairs**, making them excellent for **quick retrievals in unstructured documents, but are too limited to work with large amounts of tabular data.**
* Arrays are easy to **create and excellent for completing tasks that involve working with sequential data but searching, sorting, inserting and deleting will pose items once you shift items.**
